```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
```

```python
# Study data files
# Create a file path
filepath = "Resources/storms_final.csv"
# Read in the data. 
storm_df = pd.read_csv(filepath)
storm_df.head()
```

```python
storm_df.info()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

hurricane_countc1 = (storm_df['category'] == 1).sum()
hurricane_countc2 = (storm_df['category'] == 2).sum()
hurricane_countc3 = (storm_df['category'] == 3).sum()
hurricane_countc4 = (storm_df['category'] == 4).sum()
hurricane_countc5 = (storm_df['category'] == 5).sum()

categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5']
counts = [hurricane_countc1, hurricane_countc2, hurricane_countc3, hurricane_countc4, hurricane_countc5]

plt.bar(categories, counts, color=['#9D0208', '#FFA500', '#FFFF00', '#FF6347', '#8B4513'])
plt.xlabel("Hurricane Category")
plt.ylabel("Number of Hurricanes")
plt.title("Total Number of Hurricanes by Category")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already calculated the counts for each hurricane category

years = storm_df['year'].unique()  # Assuming 'year' is the column representing years

plt.figure(figsize=(12, 6))

for category in range(1, 6):
    category_counts = storm_df[storm_df['category'] == category].groupby('year').size()
    plt.plot(category_counts.index, category_counts.values, label=f"Category {category}")

plt.xlabel("Years")
plt.ylabel("Number of Hurricanes")
plt.title("Number of Hurricanes by Category over the Years")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.legend()
plt.show()
```
